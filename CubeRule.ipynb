```python
import  numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from scipy import stats
import os
```

```python
os.chdir(r'C:\Users\user\Desktop\uoc\Applied Statistics\Pr2')

df = pd.read_csv(r'state_legislative_election_results_1971_20121.csv', low_memory = False)
```

```python
years = np.array([i for i in range(1972,2013,2)])  #generates an array for the years we wish to test
```

```python
# Creates the two arrays needed for the hypothesis testing: The analogy of the total votes for all states per year  
# between Democrats and Republicans and the analogy of the seats in cube root

votes = []
mps = []

for i in years:
    ALL = df[(df['Year'] == i) & (df['Incumbent'] == 1)]
    
    D_votes = ALL['Dem Votes'].sum()
    R_votes = ALL['GOP Votes'].sum()

    Dems = ALL[ALL['Party'] == 'D'].groupby('Party').sum().iloc[0]['Incumbent']
    Reps = ALL[ALL['Party'] == 'R'].groupby('Party').sum().iloc[0]['Incumbent']
    
    votes.append(D_votes/R_votes)
    mps.append((Dems/Reps)**(1/3))
    
```

```python
# Bar chart of the two fractions per year

x = np.ones(len(years))*0.5

green_patch = mpatches.Patch(color='green', label='Seats (Cube rooted)')
orange_patch = mpatches.Patch(color='orange', label='Votes')

fig = plt.figure(figsize = (16,6))
ax = fig.add_axes([0,0,1,1])
ax.bar(years,mps,width = 0.25, color = 'g')
ax.bar(years+x,votes, width = 0.25, color = 'orange')
ax.set_title("Bar Chart: Fraction of Dems to Gop votes vs Fraction of Dems to Gop seats")
ax.set_ylabel("Fractions")
ax.set_xlabel("Years")
ax.legend(handles = [green_patch, orange_patch])
plt.show()
```

```python
#Creating the histogram for the the binned Kolmogorov-Smirnov test

hist_votes, bin_edges_votes = np.histogram(votes,bins = 5,range = (min(votes), max(votes)))
binned_votes_centers = 0.5*(bin_edges_votes[1:] + bin_edges_votes[:-1])
hist_mps, bin_edges_mps = np.histogram(mps,5,range = (min(mps), max(mps)))
binned_mps_centers = 0.5*(bin_edges_votes[1:] + bin_edges_votes[:-1])
```

```python
#Running the test and returning the p-value

KS = stats.ks_2samp(np.repeat(binned_votes_centers,hist_votes),np.repeat(binned_mps_centers, hist_mps))

print ("The binned Kolmogorov-Smirnov test returns p value = {} ".format(KS[1]))
```

```python
# Histogram of the two distributions

blue_patch = mpatches.Patch(color='blue', label='Distribution of Seats (Cube rooted)')
orange_patch = mpatches.Patch(color='orange', label=' Distribution of Votes')

fig = plt.figure(figsize = (10,6))
ax = fig.add_axes([0,0,1,1])
ax.hist([votes,mps], bins = 5)
ax.set_title(" Distribution of Votes vs Distribution of Seats")
ax.legend(handles = [blue_patch, orange_patch])

plt.show()
```
